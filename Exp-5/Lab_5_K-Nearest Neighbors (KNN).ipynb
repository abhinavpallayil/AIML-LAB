```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X,y=make_blobs(n_samples=500,
               n_features=2,centers=4,
               cluster_std=1.5,random_state=4)
```

```python
plt.style.use('seaborn')
plt.figure()
plt.scatter(X[:,0], X[:,1], c=y, marker= '*',s=100,edgecolors='black')
plt.title("Input Data", fontsize=15)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
k_value = int(input("Enter k value: "))
```

```python
%%capture --no-display
knn_value = KNeighborsClassifier(n_neighbors = k_value)
knn_value.fit(X_train, y_train)
y_pred_value = knn_value.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy of this model with k = %d is %.2f%%'
      %(k_value,accuracy_score(y_test, y_pred_value)*100))
```

```python
plt.figure()
plt.scatter(X_test[:,0],X_test[:,1],
            c=y_pred_value,marker='*',s=100,edgecolors='black')
plt.title(f"KNN classifier predicted values with k = %d"%k_value,
          fontsize=15)
```
