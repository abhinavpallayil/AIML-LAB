```python
import numpy as np
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
N = 1000
def makeData(x):
    r = [a/10 for a in x]
    y = np.random.uniform(-.5, .2, len(x))
    return np.array(y+r)
x = [i/100 for i in range(N)]
y = makeData(x)
x = np.array(x).reshape(-1,1)
```

```python
plt.scatter(x, y, s=5, color="blue")
plt.title('Linear data')
plt.show()
```

```python
svr = SVR().fit(x, y)
yfit = svr.predict(x)

plt.scatter(x, y, s=5, color="blue", label="original")
plt.plot(x, yfit, lw=2, color="red", label="fitted")
plt.legend()
plt.title('SVR classifier')
plt.show()
```

```python
score = svr.score(x,y)
print("R-squared: %.4f" %score)
print("MSE: %.4f"%mean_squared_error(y, yfit))
```
